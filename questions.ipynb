Since you've tackled the "Rite of Passage" of basic percentages and systems of equations, here are four problems with a similar "logic translation" difficulty, but with different real-world flavors.

### 1. The Warehouse Inventory (System of Equations)
A tech warehouse stocks two types of tablets: **Basic** ($300) and **Pro** ($800). 
* The morning inventory count shows **120 total tablets** on the shelves.
* The insurance value of the entire tablet inventory is **$\$56,000$**.
* **Task**: Write a script to calculate exactly how many Basic and Pro tablets are in stock.

### 2. The Subscription Churn (Percentage Logic)
An app started the year with **25,000 active subscribers**. 
* In the first six months, they lost **15%** of their users (churn).
* In the second half of the year, they ran a marketing campaign and gained **22%** more users (based on the number they had at the 6-month mark).
* **Task**: Calculate the final subscriber count and determine if they ended the year with more or fewer users than they started with.

### 3. The Server Efficiency (Time & Rate)
A data center has a "Job Queue" of **1,200 tasks**. 
* **Server A** can process **40 tasks per hour**.
* **Server B** is **25% faster** than Server A.
* **Task**: If both servers work together, how many hours will it take to clear the entire queue? 

### 4. The Currency Fluctuation (Comparative Ratios)
You are building a travel app.
* **Last year**: $1$ USD was worth $1.32$ CAD.
* **This year**: $1$ USD is worth $1.38$ CAD.
* A hotel room in Toronto costs **$250 CAD** per night.
* **Task**: Calculate the percentage change in the cost of the room for an American traveler (paying in USD). Did the room get cheaper or more expensive for them?

---

## Why these are "Different Flavors"
* **Inventory**: Same logic as the ticket problem (System of Equations) but applies to asset management.
* **Churn**: This tests "Compounding Percentages." You cannot simply add $-15\% + 22\%$. You must apply them sequentially.
* **Server**: This introduces "Rates." It’s similar to the "Hours in a Week" problem but requires combining speeds.
* **Currency**: This is a "Double-Step Percentage" problem. You have to convert, then compare.
