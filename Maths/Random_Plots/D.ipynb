```python
import math

def discriminant():

    a = int(input("Enter a: "))
    b = int(input("Enter b: "))
    c = int(input("Enter c: "))

    d = b**2 - 4*a*c
    print("Discriminant =", d)

    if d < 0:
        print("It will have two complex roots")

    elif d == 0:
        print("It will have one real root")

    # 🔧 Case 3: positive → need square test
    else:
        # integer square root
        root = math.isqrt(d)

        # ✅ perfect square test
        if root * root == d:
            print("It will have two rational roots")
        else:
            print("It will have two irrational roots")
```
