```python
import matplotlib.pyplot as plt
import scipy as sp

def take_plots():
    x_vals = []
    y_vals = []

    for _ in range(10):
        x = float(input("Enter x: "))
        y = float(input("Enter y: "))
        x_vals.append(x)
        y_vals.append(y)

    plt.plot(x_vals, y_vals, marker='o')
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.show()

take_plots()
```
