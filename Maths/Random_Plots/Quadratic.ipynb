```python
# Second
import math

def quadratic(equation):

    parts = equation.split("+")

    # 🔧 clean spaces
    first = parts[0].strip()
    second = parts[1].strip()
    third = parts[2].strip()

    # 🔧 extract a
    a = first.replace("x^2", "")
    if a == "":
        a = 1
    a = int(a)

    # 🔧 extract b
    b = second.replace("x", "")
    if b == "":
        b = 1
    b = int(b)

    # 🔧 extract c
    c = int(third)

    print("a =", a, "b =", b, "c =", c)

    # 🔧 discriminant
    D = b*b - 4*a*c
    print("Discriminant =", D)

    if D < 0:
        print("No real roots")
        return

    # 🔧 quadratic formula
    r1 = (-b + math.sqrt(D)) / (2*a)
    r2 = (-b - math.sqrt(D)) / (2*a)

    print("root 1 =", r1)
    print("root 2 =", r2)

print(quadratic("56x^2 + 44x + 23"))
```
