```python
import matplotlib.pyplot as plt

n_values = []
e_approx = []

for n in range(1,100):
    value = (1+1/n)**n
    n_values.append(n)
    e_approx.append(value)

plt.plot(n_values, e_approx)
plt.show
```

```python
# Using numpy
import numpy as np

n = np.arange(1,101)
e_approx1 = (1+1/n)**n

plt.plot(n, e_approx1)
plt.show()
```
