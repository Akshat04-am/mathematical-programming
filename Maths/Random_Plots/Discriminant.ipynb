```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define your coefficients (Change these numbers to test different graphs!)
a = 1
b = -5
c = 6

# 2. Calculate the Discriminant
discriminant = b**2 - 4*a*c

# 3. Use "if" statements to check the nature of the roots
print(f"The Discriminant is: {discriminant}")

if discriminant > 0:
    print("Nature: Two distinct real roots.")
elif discriminant == 0:
    print("Nature: One real root (Vertex touches x-axis).")
else:
    print("Nature: Complex (Imaginary) roots.")

# 4. Visualization (Let's see the Parabola)
# Create a range of x values around the center
x = np.linspace(-10, 10, 400)
y = a*x**2 + b*x + c

plt.axhline(0, color='black', linewidth=1) # The x-axis
plt.axvline(0, color='black', linewidth=1) # The y-axis
plt.plot(x, y, label=f'{a}x² + {b}x + {c}')
plt.grid(True, linestyle='--')
plt.legend()
plt.title("Quadratic Function Visualization")
plt.show()
```
