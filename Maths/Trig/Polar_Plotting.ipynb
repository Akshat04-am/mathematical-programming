### Make a automatic script to take a equation and plot polar points for key trig ratios

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup the data
theta = np.linspace(0, np.pi, 500)
r = 2 * np.sin(theta)

# Specific points from Figure 4
pts_theta = [0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6]
pts_r = [0, 1, 1.73, 2, 1.73, 1]

# 2. Create the plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot the continuous line
ax.plot(theta, r, color='tab:blue', linewidth=2)

# Plot the specific markers
ax.scatter(pts_theta, pts_r, color='black', zorder=5)

# 3. Visual Formatting
ax.set_ylim(0, 2.5)  # Set radial limit
ax.set_rticks([1, 2]) # Radial grid lines
ax.set_theta_zero_location('E') # Standard polar orientation
ax.grid(True)

plt.show()
```
