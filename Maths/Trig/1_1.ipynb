```python
import numpy as np

# 1. Setup the data once (Vectorized)
degrees = np.arange(0, 361, 15)
radians = np.radians(degrees)
sines = np.sin(radians)
cosines = np.cos(radians)

# 2. Print the table
print(f"{'Deg':<5} | {'Sin':<8} | {'Cos':<8}")
print("-" * 25)

for i in range(len(degrees)):
    print(f"{degrees[i]:<5} | {sines[i]:<8.4f} | {cosines[i]:<8.4f}")
```
