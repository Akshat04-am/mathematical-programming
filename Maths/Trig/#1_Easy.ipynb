```python
# 1. Radians to degree converter.
import numpy as np

degrees = 45
# Manual algebraic way: deg * (pi / 180)
radians_manual = degrees * (np.pi / 180)

# Built-in NumPy way
radians_np = np.radians(degrees)

print(f"{degrees} degrees is {radians_np:.4f} radians")
```

```python
# 2. Degree + hypotenuse = cos and sin 
import numpy as np

r = 10         # Distance from center
theta = np.radians(30) # 30 degree angle

cos = r * np.cos(theta)
sin = r * np.sin(theta)

print(f"Coordinates: ({cos:.2f}, {sin:.2f})")
print(f"Sum of {cos.round(2)} + {sin.round(2)} is {(cos**2) + (sin**2)}")
```

```python
# 3. Finding the Angle (Inverse Trig)
import numpy as np

cos, sin = -5, 5

# Get angle in radians
angle_rad = np.arctan2(sin, cos)

# Convert back to degrees for readability
angle_deg = np.degrees(angle_rad)

print(f"The angle to point ({cos}, {sin}) is {angle_deg} degrees")
```

```python
# 4. Simple Visualization (Sine Wave)
import numpy as np
import matplotlib.pyplot as plt

# 100 points from 0 to 360 degrees (converted to radians)
angles = np.linspace(0,2 * np.pi, 50)
amplitudes = np.sin(angles)

plt.plot(angles, amplitudes)
plt.axhline(0, color='black', lw=1) # Baseline
plt.show()
```
