###  Degree Table: 
 **Create** a loop that prints the sin and cos values for every 15 degrees between 0 and 90.

```python
import numpy as np

# Define the degrees using a range (0 to 90 inclusive, step 15)
degrees = np.arange(0, 91, 15)

print(f"{'Degrees':<10} | {'Sine':<10} | {'Cosine':<10}")
print("-" * 35)

for deg in degrees:
    rad = np.radians(deg)
    sin = np.sin(rad)
    cos = np.cos(rad)
    
    print(f"{deg:<10} | {sin:<10.4f} | {cos:<10.4f}")
```

### Cos and Sin
**Alternate** between cos and sin for every 15 degrees. 

```python
import numpy as np

degrees = np.arange(0, 361, 15)

print(f"{'Degrees':<10} | {'Func':<5} | {'Value':<10}")
print("-" * 30)

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    
    # Alternating logic using modulo on index i
    if i % 2 == 0:
        f_type = "sin"
        val = np.sin(rad)
    else:
        f_type = "cos"
        val = np.cos(rad)
    
    print(f"{deg:<10} | {f_type:<5} | {val:<10.4f}")
```

### Graph the above

```python
import numpy as np
import matplotlib.pyplot as plt

degrees = np.arange(0, 361, 15)
values = []
labels = []

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    if i % 2 == 0:
        values.append(np.sin(rad))
        labels.append('sin')
    else:
        values.append(np.cos(rad))
        labels.append('cos')

plt.plot(degrees, values, marker='o', linestyle='--')
plt.title("Alternating Sin/Cos Every 15 Degrees")
plt.xlabel("Degrees")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

### The Coordinate Unpacker
**Coords** = `[[0, 1], [30, 0.5], [90, 1], [180, 0]]`. 

**Task**: Loop through coords. In the loop header, use multiple assignment to unpack the pair into deg, expected_val.

**Logic:** Calculate actual_val = np.sin(np.radians(deg)). Use an if to check if actual_val is "Close enough" to expected_val.

**Check**: Add a check for "mathematical impossibility": if deg > 360, print an error and continue.

```python
import numpy as np

# Your two separate lists
degrees = [0, 30, 90, 180, 400]
expected = [0, 0.5, 1, 0, 0]

# zip() pairs them: (0, 0), (30, 0.5), (90, 1)...
for deg, exp in zip(degrees, expected):
    
    # 1. Check for mathematical impossibility first
    if deg > 360:
        print(f"Error: {deg} is out of bounds.")
        continue
    
    # 2. Perform the actual math
    actual = np.sin(np.radians(deg))
    
    # 3. Use Nested Logic to check the result
    if np.isclose(actual, exp, atol=1e-5):
        print(f"Deg {deg}: Match.")
    else:
        print(f"Deg {deg}: Mismatch. Expected {exp}, got {actual:.2f}")
```
