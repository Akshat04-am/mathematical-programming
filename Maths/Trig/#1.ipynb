### 1. The Coordinate Unpacker
* ** You have a list of coordinate pairs: coords = `[[0, 1], [30, 0.5], [90, 1], [180, 0]]`. **

**Task**: Loop through coords. In the loop header, use multiple assignment to unpack the pair into deg, expected_val.

**Logic:** Calculate actual_val = np.sin(np.radians(deg)). Use an if to check if actual_val is "Close enough" to expected_val.

**Check**: Add a check for "mathematical impossibility": if deg > 360, print an error and continue.

```python
import numpy as np

# This is your data structure: [degree, expected_value]
coords = [[0, 0.0], [30, 0.5], [90, 1.0], [180, 0.0], [400, 0.0]]

for deg, expected in coords:
    # 1. Boundary Check (Mathematical Impossibility)
    if deg > 360:
        print(f"Error: {deg} is out of bounds (Max 360).")
        continue # Skips the rest of this turn and moves to the next pair
        
    # 2. Math Logic
    actual = np.sin(np.radians(deg))
    
    # 3. Nested Logic: Compare actual vs expected
    # We use np.isclose because 0.0000000001 is basically 0
    if np.isclose(actual, expected, atol=1e-5):
        print(f"Deg {deg:>3}: Match! ({actual:.2f})")
    else:
        print(f"Deg {deg:>3}: Logic Error. Expected {expected}, got {actual:.2f}")
```

```python
import numpy as np

# Your two separate lists
degrees = [0, 30, 90, 180, 400]
expected = [0, 0.5, 1, 0, 0]

# zip() pairs them: (0, 0), (30, 0.5), (90, 1)...
for deg, exp in zip(degrees, expected):
    
    # 1. Check for mathematical impossibility first
    if deg > 360:
        print(f"Error: {deg} is out of bounds.")
        continue
    
    # 2. Perform the actual math
    actual = np.sin(np.radians(deg))
    
    # 3. Use Nested Logic to check the result
    if np.isclose(actual, exp, atol=1e-5):
        print(f"Deg {deg}: Match.")
    else:
        print(f"Deg {deg}: Mismatch. Expected {exp}, got {actual:.2f}")
```

### The "Validation Engine":

* **data** =`'[[0, 0], [45, 0.7071], [90, 1.0], [400, "SKIP"]]`

Write a loop that:
Unpacks the degree and the expected_sin.

**Check 1**: If the expected_sin is the string "SKIP", use the continue keyword to ignore it.

**Check 2:** If the degree is $> 360$, print "Invalid Angle" and break the entire loop (kill the motor).

Math: Calculate the actual_sin.

Verification: Use np.isclose with an atol=1e-3 to see if the motor is on track. If it matches, print "Pass", otherwise print "Fail".

```python
import numpy as np

data = [[0, 0], [45, 0.7071], [90, 1.0], [400, "SKIP"]]

for degree,expected_sin in data:
    if expected_sin == "SKIP":
        continue
    elif degree > 360:
        print("invalid Angle")
        break

    actual = np.sin(np.radians(degree))

    if np.isclose(actual, expected_sin, atol=1e-5):
        print(f"Deg {degree}: Match.")
    else:
        print(f"Deg {degree}: Mismatch. Expected {expected_sin}, got {actual:.2f}")
```
