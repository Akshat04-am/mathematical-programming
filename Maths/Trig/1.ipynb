 Degree Table: Create a loop that prints the sin and cos values for every 15 degrees between 0 and 90.

```python
import numpy as np

# Define the degrees using a range (0 to 90 inclusive, step 15)
degrees = np.arange(0, 91, 15)

print(f"{'Degrees':<10} | {'Sine':<10} | {'Cosine':<10}")
print("-" * 35)

for deg in degrees:
    rad = np.radians(deg)
    s_val = np.sin(rad)
    c_val = np.cos(rad)
    
    print(f"{deg:<10} | {s_val:<10.4f} | {c_val:<10.4f}")
```

```python
import numpy as np

degrees = np.arange(0, 361, 15)

print(f"{'Degrees':<10} | {'Func':<5} | {'Value':<10}")
print("-" * 30)

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    
    # Alternating logic using modulo on index i
    if i % 2 == 0:
        f_type = "sin"
        val = np.sin(rad)
    else:
        f_type = "cos"
        val = np.cos(rad)
    
    print(f"{deg:<10} | {f_type:<5} | {val:<10.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

degrees = np.arange(0, 361, 15)
values = []
labels = []

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    if i % 2 == 0:
        values.append(np.sin(rad))
        labels.append('sin')
    else:
        values.append(np.cos(rad))
        labels.append('cos')

plt.plot(degrees, values, marker='o', linestyle='--')
plt.title("Alternating Sin/Cos Every 15 Degrees")
plt.xlabel("Degrees")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
