### Making a circle.

```python
# Making a circle.
import numpy as np
import matplotlib.pyplot as plt

# 1. linspace = Number of points (num).
theta = np.linspace(0, 2 * np.pi, 100)

# 2. Cosine handles horizontal and Sine vertical.
x = np.cos(theta)
y = np.sin(theta)

plt.plot(x, y)
plt.gca().set_aspect('equal') # Prevents distortion
plt.show()
```

### Mapping trig angles to 2D coordinates.

```python
import numpy as np
angle = np.radians(60) #Use np.radians()

x = np.cos(angle)
y = np.sin(angle)

# Output logic
print(f"X (Cos) is {round(x, 4)}") # Should be 0
print(f"Y (Sin) is {round(y, 4)}") # Should be 1
```

### Calculating sin and cos from theta and hypotenuse

```python
# cos,sin = adj,height/hyp.
import numpy as np

r = 10         # Distance from center
theta = np.radians(30) # 30 degree angle

cos = r * np.cos(theta)
sin = r * np.sin(theta)

print(f"Coordinates: ({cos:.2f}, {sin:.2f})")
print(f"Sum of {cos.round(2)} + {sin.round(2)} is {(cos**2) + (sin**2)}")
```

### Calculating Tan from Sin and Cos

```python
# Sin and cos units.
cos, sin = 1 , np.sqrt(3)

# Reverse tan.
angle_rad = np.arctan2(sin, cos)

# Takes angle gives in radians.
angle_deg = np.degrees(angle_rad)

print(f"The angel to point ({cos}, {sin} is {angle_deg.round(2)} degress)")

print(angle_rad) # Radians
print(angle_deg.round(2)) # Angle
```

### Plotting Sine Wave

```python
# Simple Visualization (Sine Wave)
import numpy as np
import matplotlib.pyplot as plt

# 50 points from 0 to 360 degrees (converted to radians)
angles = np.linspace(0,2 * np.pi, 50)
amplitudes = np.sin(angles)

plt.plot(angles, amplitudes)
plt.axhline(0, color='black', lw=1) # Baseline
plt.show()
```
