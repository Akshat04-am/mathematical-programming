```python
# Plotting point intercept
import matplotlib.pyplot as plt

def point_intercept(x1, y1, x2, y2):

    # Calculating slope.
    m = (y2 - y1) / (x2 - x1)

    # Linear Eq. (y = mx + b)
    b = y1 - m * x1

    # Modification for vertical line.
    if x1 == x2:
        print(f"x1 cant be equal to x2")
        return

    x_plot = [x1, x2]
    y_plot = [y1, y2]

    plt.plot(x_plot, y_plot,  'r')
    plt.axhline(0, color = 'black')
    plt.axvline(0, color = 'black')

    # Sets the x-axis and y-axis.
    plt.xlim(2000, 2020)
    plt.ylim(50000, 80000)

    plt.show()

    print(f"y = {m}x + {b}")

point_intercept(2010, 55000 , 2012, 76000)
```

```python
# Predicting further.
import matplotlib.pyplot as plt

def point_intercept(x1, y1, x2, y2, predict_x):

    # Calculating slope.
    m = (y2 - y1) / (x2 - x1)

    # Linear Eq. (y = mx + b)
    b = y1 - m * x1

    # Modification for vertical line.
    if x1 == x2:
        print(f"x1 cant be equal to x2")
        return

    x_plot = [x1, x2]
    y_plot = [y1, y2]

    my_pred = m * predict_x + b

    plt.plot(x_plot, y_plot,  'r')
    plt.axhline(0, color = 'black')
    plt.axvline(0, color = 'black')

    # Sets the x-axis and y-axis.
    plt.xlim(2000, 2020)
    plt.ylim(50000, 80000)

    plt.show()

    print(f"y = {m}x + {b}")
    print(my_pred)

point_intercept(2010, 55000 , 2012, 76000, 2016)
```
