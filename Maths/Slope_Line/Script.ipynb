```python
import matplotlib.pyplot as plt
import numpy as np

xmin, xmax, ymin, ymax = -10, 10, -10, 10

# Create data arrays
x_values = np.arange(xmin, xmax + 1)
y_values = 0.5 * x_values + 1

# Forces to stay visually
fig, ax = plt.subplots()
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

# Draw axes
ax.axhline(0, color='blue')
ax.axvline(0, color='blue')

# Plot all points at once
plt.plot(x_values, y_values, "ro") 

# Print table
# print("x \t y")
# for x, y in zip(x_values, y_values):
#     print(f"{x} \t {y}")

plt.show()
```

### Function

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_graph(xmin=-10, xmax=10, ymin=-10, ymax=10):
    # Establishing Plot
    fig, ax = plt.subplots()

    # Locks in range.
    ax.set_xlim([xmin, xmax])
    ax.set_ylim([ymin, ymax])

    # Sets vertical and horizontal line.
    ax.axhline(0, color="blue")
    ax.axvline(0, color="blue")
    
    # Grid for better visibility.
    ax.grid(True, linestyle='--', alpha=0.6)
    
    return fig, ax
```

```python
# Controls numbers on x and y axis
ax.set_xticks(np.arange(xmin, xmax))
ax.set_yticks(np.arange(ymin, ymax))
```

### Examples

```python
#Random slope intercept eq
fig,ax = draw_graph()

x = np.arange(-10, 11)
y = 0.5 * x + 1

ax.plot(x, y, "ro")
plt.show()
```

```python
# Two separate graphs (ax1 and ax2)
fig,(ax1, ax2) = plt.subplots(1, 2)

# Connected ONLY to the first graph
ax1.plot(x, y, "ro") 

# Connected ONLY to the second graph
ax2.plot(x, y, "bo")
plt.show()
```
