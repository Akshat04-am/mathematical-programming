### Divide out any perfect square factors. 
**For example:<br>$\sqrt{12}$ = $\sqrt{4 * 3}$ = 2$\sqrt{3}$<br>Because 4 is a square number, the square root of 4 is now outside the radical.**

```python
import math

# number to factor
n = 12

# The largest possible integer whose square can be a factor of 'number'
upper_limit = math.floor(math.sqrt(n)) + 1

max_factor = 1
for maybe_factor in range(1,upper_limit):
    if n % (maybe_factor**2) == 0:
        max_factor = maybe_factor

print("n = ", n) 
print("Square rooted factor = ", max_factor) 
print("Square factor = ", max_factor**2) 
print("integer: ", n/(max_factor**2)) 
```

### Using Sympy to make this look nice

```python
import math
import sympy
from sympy import symbols

n = 96

upper_limit = math.floor(math.sqrt(n)) + 1
max_factor = 1
other_factor = 1
square_root = 1

for maybe_factor in range(1, upper_limit):
    if n % (maybe_factor**2) == 0:
        max_factor = maybe_factor**2

other_factor = n/max_factor

square_root = int(math.sqrt(max_factor))
other_factor = int(other_factor)
output = square_root*sympy.sqrt(other_factor)

print(n)
print(f"Max Factor: {max_factor}")
print(f"Sqrt: {square_root} | {square_root**2}")
print(f"Other_Factor: {other_factor}")

# The Output
output
```
