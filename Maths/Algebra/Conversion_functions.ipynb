```python
def dec_to_frac(in_string):
    parts = in_string.split(".")

    n = parts[0]
    d = parts[1]

    exponents = len(d)
    new_n = n * 10**exponents
    new_d = d*10**exponents
    frac = new_n/new_d
    return frac


print(dec_to_frac("3.333"))
```

```python
from math import gcd

def dec_to_frac(in_string):
    parts = in_string.split(".")

    whole = int(parts[0])
    decimal = int(parts[1])

    power = len(parts[1])

    numerator = whole * (10 ** power) + decimal
    denominator = 10 ** power

    g = gcd(numerator, denominator)

    numerator = numerator // g          
    denominator = denominator // g  

    return numerator, denominator

print(dec_to_frac("1.25"))
```

```python
def frac_to_perc(in_string):
    parts = in_string.split("/")

    n = int(parts[0])
    d = int(parts[1])

    value  = n/d
    percent = value * 100

    return percent

print(frac_to_perc("1/4"))
```

```python
import sympy as sp

a, b, c, x = sp.symbols("a b c x")

expr = a*x**2 + b*x + c

solution = sp.solve(expr, x)

for s in solution:
    print(f"x = {s}")
```

```python
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans
```
