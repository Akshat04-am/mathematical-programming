**Scenario: The London Eye (Modified for Logic Translation)**
A large observation wheel has a diameter of **120 meters** and sits on a platform such that the lowest point of the wheel is **5 meters** above the ground. The wheel completes one full revolution every **30 minutes**. A passenger boards at the lowest point at $t = 0$.



**Computational Requirements:**
* **Model the Height:** Define the function $h(t)$ that calculates the passenger's height above the ground at any time $t$ (in minutes). 
    * Use the general form: $$h(t) = A \cos(B(t - C)) + D$$
* **Task 1 (Logic):** Calculate the constants $A, B, C, \text{ and } D$ within your script based on the physical dimensions provided.
* **Task 2 (Calculation):** Write the code to output the height at $t = 5, 10, \text{ and } 15$ minutes.
* **Task 3 (Equation Solving):** Find the first time $t > 0$ when the passenger is exactly **95 meters** above the ground. 

---

```python
import math
import sympy as sp


diameter = 120
bottom_height = 5
period = 30
target_h = 95

radius = diameter / 2
D = bottom_height + radius
A = -radius
B = (2 * math.pi) / period
C = 0

def get_height(t):
    return A * math.cos(B * (t - C)) + D

times = [5, 10, 15]
for t in times:
    print(f"Height at {t} min: {get_height(t):.2f}m")

# -----------------
t_sym = sp.symbols('t')
B_sym = (2 * sp.pi) / period
equation = sp.Eq(A * sp.cos(B_sym * (t_sym - C)) +D, target_h)

solution = sp.solve(equation, t_sym)
print(round(solution[0]))
```
