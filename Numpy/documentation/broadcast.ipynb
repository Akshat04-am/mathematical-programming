```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Add a vector to a matrix
A = np.arange(9).reshape(3,3)
a = np.array([100,200,300])

print("A:\n", A)
print("a:", a)
print("A + a:\n", A + a)
```

```python
# 2. Add a column vector to a matrix
B = np.arange(9).reshape(3,3)
b = np.array([[10],[20],[30]])

print(B)
```

```python
print("B:\n", B)
print("b:\n", b)
print("B + b:\n", B + b)
```

```python
# 3. Multiply a matrix by a 1-D array
C = np.random.randint(1,5,size=(4,4))
c = np.array([1,10,100,1000])

print("C * c:\n", C * c)
```

```python
# 4. Add a scalar
D = np.random.randint(-5,5, size=(3,5))
print(D + 7)
```

```python
# 5. Add two shapes that should fail (see the error)
X = np.ones((3,4))
Y = np.ones((2,3))
X + Y
```

```python
# 6. Broadcast a 1×3 row across a 3×3 matrix
A = np.arange(9).reshape(3,3)
v = np.array([[10,20,30]])

print(A + v)
```

```python
# 7. Broadcast a 3×1 column across a 3×3 matrix
A = np.arange(9).reshape(3,3)
v = np.array([[10],[20],[30]])

print(A + v)
```

```python
# 8. Multiply 1-D array by 2-D matrix
A = np.arange(12).reshape(4,3)
w = np.array([100,200,300])

print(A * w)
```

```python
# 9. Try a crazy shape pairing and see what breaks
X = np.random.randn(2,3,4)
Y = np.random.randn(4)

print(X + Y)
```
