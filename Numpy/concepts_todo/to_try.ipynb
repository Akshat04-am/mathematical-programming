```python
import numpy as np

# assume prices: rows = days, cols = tickers
prices = np.array([
 [100, 50],
 [101, 49],
 [103, 50],
 [102, 52]
], dtype=float)

# 1) simple returns
returns = prices[1:] / prices[:-1] - 1  # elementwise
# 2) log returns
logret = np.log(prices[1:] / prices[:-1])

# 3) mean & std (daily)
mean = returns.mean(axis=0)
std = returns.std(axis=0, ddof=0)

print(returns, logret, mean, std)
```

```python
# daily returns from above
weights = np.array([0.6, 0.4])
daily_portfolio = returns.dot(weights)                # portfolio daily returns
cov = returns.T.dot(returns) / (returns.shape[0])    # population covariance
port_var = weights.T.dot(cov).dot(weights)
port_std = np.sqrt(port_var)

print("Daily port returns:", daily_portfolio)
print("Portfolio std:", port_std)
```

```python
# use mean,std from returns, simulate 10000 daily returns
import numpy as np
mu = mean.mean()
sigma = std.mean()
sim = np.random.normal(mu, sigma, size=10000)
prob_loss = np.mean(sim < 0)
prob_loss
```

```python
# Covariance matrix (3x3)
cov_matrix = np.cov(returns)

# Portfolio weights
weights = np.array([0.4, 0.3, 0.3])

# Portfolio variance = wᵀ Σ w
port_variance = weights.T @ cov_matrix @ weights

# Portfolio std dev (risk)
port_std = np.sqrt(port_variance)

# Expected portfolio return
port_return = np.dot(weights, mean_returns)
```
