```python
import sympy as sp

x, y = sp.symbols('x y')

# Equation 1: Headcount
eq1 = sp.Eq(x + y, 5332)
# Equation 2: Revenue
eq2 = sp.Eq(20*x + 10*y, 71750)

solution = sp.solve((eq1, eq2), (x, y))
print(solution) 
# Output: {x: 1843, y: 3489}
```

```python
total_attend = 5332
total_revenue = 71750

expensive_price = 20
cheaper_price = 10

# Logic: Loop through all possible expensive ticket counts
for expensive_count in range(total_attend + 1):
    # If x people bought expensive, then (total - x) must be cheaper
    cheaper_count = total_attend - expensive_count
    
    current_total = (expensive_count * expensive_price) + (cheaper_count * cheaper_price)
    
    if current_total == total_revenue:
        print(f"Expensive tickets sold: {expensive_count}")
        print(f"Cheaper tickets sold: {cheaper_count}")
        break
```
