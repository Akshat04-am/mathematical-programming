### **Script is doing.**
1. Finding Intersection Points (System Solving)
2. Isosating $y$ (Function Creation)
3. Extraction and Visualization

```python
import sympy as sp
from sympy.plotting import plot

x, y = sp.symbols("x y")
f1 = -x**2 - y + 10
f2 = 2*x**2 - 2*y - 4

# 1. Solve for intersection points
solution = sp.nonlinsolve([f1, f2], (x, y))

for x_val, y_val in solution:
    print(f"Intersection: X = {x_val}, Y = {y_val}")

# 2. Isolate y for plotting
y1 = sp.solve(sp.Eq(f1, 0), y) # Returns [10 - x**2]
y2 = sp.solve(sp.Eq(f2, 0), y) # Returns [x**2 - 2]

print("Equation 1: y =", y1[0])
print("Equation 2: y =", y2[0])

# 3. Plotting
xmin, xmax = -5, 5
plot(y1[0], y2[0], (x, xmin, xmax))
```
