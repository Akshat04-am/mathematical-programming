Nested Logic: Keep your current sin/cos toggle, but add an extra check: if the val is exactly 1.0 or -1.0, print a message next to it saying "PEAK".

The Manual Toggle: Instead of using the index i and modulo %, create a variable is_sin = True before the loop. Inside the loop, use if is_sin:. At the very bottom of the loop, write a line that flips the boolean (is_sin = not is_sin).

```python
import numpy as np

degrees = np.arange(0, 361, 15)

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    
    # 1. First Layer: Toggle the function based on index
    if i % 2 == 0:
        f_type = "sin"
        val = np.sin(rad)
    else:
        f_type = "cos"
        val = np.cos(rad)
    
    # Clean up floating point noise (e.g., 0.999999999 -> 1.0)
    val = np.round(val, 4)
    
    # 2. Nested Layer: Check the VALUE for a "Peak"
    # This happens regardless of whether it's sin or cos
    message = ""
    if val == 1.0 or val == -1.0:
        message = " <--- PEAK"
    
    print(f"{deg:<10} | {f_type:<5} | {val:<10.4f}{message}")
```

```python
import numpy as np

degrees = np.arange(0, 361, 15)

# enumerate gives us 'i' for the table structure and 'deg' for the math
for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    
    # Use 'i' (Index) to control the alternating pattern
    if i % 2 == 0:
        f_type, val = "sin", np.sin(rad)
    else:
        f_type, val = "cos", np.cos(rad)
    
    # Use 'val' (Algebraic result) to check for the peak
    # Using np.isclose because floating point math is rarely exactly 1.0
    suffix = ""
    if np.isclose(np.abs(val), 1.0):
        suffix = " <--- PEAK"
        
    print(f"{deg:>3}° | {f_type} | {val:>7.4f}{suffix}")
```
