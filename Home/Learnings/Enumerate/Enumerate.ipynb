Index Multiplier: Create a list [2, 4, 6, 8]. Use a manual counter i to print the result of the current number multiplied by its index (e.g., $2 \times 0$, $4 \times 1$, etc.).

```python
import numpy as np

x = np.arange(0,101,17)

counter = 0
for val in x:

    result = val * counter
    print(f"Index {counter} | Value {val} | Result: {result}")
    counter += 1

for i, deg in enumerate(x):
    print(f"Index {i} | Value {deg}")
```

The "Even Index" Filter: Create a list of 6 random numbers. Use a manual counter to print only the numbers that sit at an even index ($0, 2, 4$).

```python
import numpy as np

# Create a list (array) of 6 random numbers
random_data = np.random.randint(0, 101, 6)

counter = 0  # Initialize manual counter
for val in random_data:
    # Logic: Check if the current index is even
    if counter % 2 == 0:
        print(f"Index {counter}: {val}")
    
    counter += 1  # Increment manual counter
```
