Nested Logic: Keep your current sin/cos toggle, but add an extra check: if the val is exactly 1.0 or -1.0, print a message next to it saying "PEAK".

```python
# Alternate bw sin and cos.
import numpy as np

degrees = np.arange(0, 361, 15)

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    
    # 1. First Layer: Toggle the function based on index
    if i % 2 == 0:
        f_type = "sin"
        val = np.sin(rad)
    else:
        f_type = "cos"
        val = np.cos(rad)
    
    # Clean up floating point noise (e.g., 0.999999999 -> 1.0)
    val = np.round(val, 4)
    
    # 2. Nested Layer: Check the VALUE for a "Peak"
    # This happens regardless of whether it's sin or cos
    message = ""
    if val == 1.0 or val == -1.0:
        message = " <--- PEAK"
    
    print(f"{deg:<10} | {f_type:<5} | {val:<10.4f}{message}")
```

### Task 1: The Geometry Toggle
You have a list of radii: `radii = [1, 2, 3, 4, 5]`.
***Logic**: Use `enumerate` and `% 2` to alternate between calculating **Area** ($\pi r^2$) and **Circumference** ($2\pi r$).
* **Labeling**: Inside the `if`, set a variable `calc_name = "Area"`. Inside the `else`, set `calc_name = "Circum"`.
* **Goal**: Print a single line at the end of the loop: `Radius: {r} | {calc_name} | {result:.2f}`.

### Task 2: The Quadratic Root Classifier
A list of coefficients $[a, b, c]$ for the equation $ax^2 + bx + c = 0$:
`coeffs = [[1, -3, 2], [1, 2, 1], [1, 0, 1]]`
* **Logic**: Unpack `a, b, c`. Calculate the discriminant $D = b^2 - 4ac$.
* **Branching**:
    * If $D > 0$, set `root_type = "Two Real"`.
    * If $D == 0$, set `root_type = "One Real"`.
    * If $D < 0$, set `root_type = "Complex"`.
* **Goal**: Print the coefficients and the `root_type` label for each row.

### Task 3: The Power/Root Unpacker (Instructional Labeling)
Data contains a value and a "command" string:
`data = [[4, "SQ"], [9, "RT"], [16, "SQ"], [25, "RT"]]`
* **Logic**: Unpack `val, cmd`.
* **Labeling**: 
    * If `cmd == "SQ"`, calculate $res = val^2$ and set `label = "Squared"`.
    * If `cmd == "RT"`, calculate $res = \sqrt{val}$ and set `label = "Rooted"`.
* **Goal**: Use the `label` variable to print: `{val} was {label} to get {res}`.



### Task 4: The Physics Trend Tracker (State Labeling)
You are tracking altitude over time: `altitudes = [0, 150, 300, 300, 250, 100]`.
* **Setup**: Before the loop, set `prev_alt = 0`.
* **Logic**: Inside the loop, calculate `diff = current_alt - prev_alt`.
* **Labeling**:
    * If `diff > 0`, `status = "CLIMBING"`.
    * Else if `diff < 0`, `status = "DESCENDING"`.
    * Else, `status = "LEVEL"`.
* **State Update**: Use multiple assignment at the end of the loop to update the "previous" value: `prev_alt = current_alt`.
* **Goal**: Print `Alt: {current_alt} | Status: {status}`.

### Why this is the "Centerpiece"
Without the `f_type` (or `status`, `label`, `calc_name`) variable, you would be forced to write your `print` statements inside every single `if` and `else` block. This is bad for math scripts because:
1.  **Redundancy**: If you want to change the print format, you have to change it in 3 or 4 places.
2.  **Logic Errors**: It makes it harder to see the actual math because it's buried under UI/print code.
3.  **Post-Processing**: Often, you need that label for more than just printing—you might need to save it to a file or use it in a calculation later in the loop.
