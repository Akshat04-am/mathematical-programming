### 1. List: The Ordered Sequence
Lists are for when order matters and you need to access things by their position.

```python
results_list = [10, 20, 30, 20]
print(results_list[0])  # Output: 10
print(results_list)     # Output: [10, 20, 30, 20] (Duplicates kept)
```

### 2. Set: The Unique Collection
Sets are for mathematical uniqueness. They don't care about order and don't allow duplicates.

```python
results_set = {10, 20, 30, 20}
# print(results_set[0]) # ERROR: Sets cannot be indexed
print(results_set)      # Output: {10, 20, 30} (Duplicates removed)
```

### 3. Dictionary: The Labeled Map
Dictionaries store data behind specific keys rather than index numbers.

```python
results_dict = {"x": 5, "y": 10}
print(results_dict["x"]) # Output: 5
```
