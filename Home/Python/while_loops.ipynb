## 1. The "Safety Check" (Basic Input Validation)
This is the most common use. You don't know how many times a user will provide bad data. You keep looping until they get it right.

**Task:** Write a script that asks for a "Probability" value. 
* A probability must be between $0$ and $1$.
* If the user enters $1.5$ or $-0.2$, the script should say "Invalid" and ask again.
* **Goal:** The loop only breaks when a valid number is entered.

```python
valid = False
while not valid:
    p = float(input("Enter probability (0 to 1): "))
    if 0 <= p <= 1:
        valid = True
        print("Accepted.")
    else:
        print("Error: Out of bounds.")
```

## 2. The "Savings Goal" (The Financial Accumulator)
This is the bread and butter of quantitative logic. You have a starting point and a target, but you don't know the "time" variable.

**Task:** * **Initial Balance:** $\$1,000$
* **Monthly Deposit:** $\$150$
* **Interest Rate:** $0.5\%$ (applied to the balance *every month* before the deposit).
* **Target:** $\$5,000$
* **Task:** Write a `while` loop to find out exactly how many **months** it takes to hit the target.

```python
initial_balance = 1000
target_balance = 5000

current_balance = initial_balance
monthly_deposit = 150
interest_rate = 0.005
month = 0

while current_balance < target_balance:
    month += 1
    current_balance *= (1 + interest_rate)
    current_balance += monthly_deposit
print(month)
```

## 3. The "Zeno’s Paradox" (Convergence Logic)
In computational math, we often find a result by getting closer and closer to it until the difference is negligible.

**Task:** * Start with `distance = 100`.
* Every step, you move **half** the remaining distance.
* **Task:** Run a loop that continues as long as `distance > 0.01`. 
* Count how many "steps" it takes to get that close to zero.

```python
distance = 100

current_distance = distance
steps = 0

while current_distance > 0.01:
    steps += 1
    current_distance /= 2
print(steps)
```

### 1. The "Inflation-Adjusted" Goal (Moving Targets)
In the previous savings problem, the target was static ($5,000$). In the real world, the target moves due to inflation.
* **Initial Balance:** $\$1,000$
* **Monthly Deposit:** $\$200$
* **Investment Return:** $1\%$ monthly.
* **Target:** Starts at $\$10,000$, but **increases by $0.5\%$** every month.
* **Task:** How many months does it take for your balance to finally "catch" the target? 
* **The Logic:** Inside the loop, you must update both the `current_balance` AND the `target_value`.

```python
initial_balance = 1000

target = 10000
target_return  = 0.005

monthly_deposit = 200
investment_return = 0.01

current_balance = initial_balance
month = 0

while current_balance < target:
    month += 1
    target *= (1 + target_return)

    current_balance *= (1+ investment_return)
    current_balance += monthly_deposit

print(f"Months to catch target: {month}")
print(f"Final Balance: {round(current_balance, 2)}")
print(f"Final Target: {round(target, 2)}")
```

### The "Bouncing Ball" Challenge (Shrinking toward a limit)
Let's apply the **`>`** logic here, because now we are **shrinking** until we hit a floor. This is a classic problem for CAT geometry and physics.

**The Setup:**
* `height = 100` (meters)
* `bounces = 0`
* **Rule:** Each bounce, the height becomes **60%** of what it was (`height *= 0.60`).
* **Condition:** We want to keep bouncing **while** the `height` is still **above 1 meter**.

**Task:** Write the script to find how many bounces it takes to drop below 1 meter.

```python
height = 100
bounces = 0

reduction = 0.60
current = height

while current > 1:
    bounces += 1

    current *= reduction

print(bounces)
```

### 2. The "Predator-Prey" Population (Variable Decay)
This models how systems collapse or stabilize. 
* **Population:** $10,000$ individuals.
* **Rule:** Every "cycle," the population grows by **10%**.
* **Constraint:** After the growth happens, **2,000** individuals are removed (the "tax" or "harvest").
* **Task:** Run a loop to see if the population hits **zero**. If it hasn't hit zero after **50 cycles**, stop the loop and print "Sustainable."
* **Critical Thinking:** If you change the removal to **1,000**, what happens to the loop?

```python
populatation = 10000
growth = 0.10
removal = 2000
month = 0

while populatation > 0:
    month += 1

    populatation *= (1 + growth)
    populatation -= removal

if month >= 50:
    print("Sustainable")
else: 
    print(f"Unsustainable: It only grew {month} months.")
```

### 3. The "Fuel Leak" (Inverse Conditions)
You are simulating a rocket burn where the mass is changing.
* **Fuel:** $500$ liters.
* **Usage:** $15$ liters per minute.
* **The Leak:** Every minute, the leak gets worse. The leak starts at **1 liter/minute** and increases by **10%** each minute.
* **Task:** How many minutes until the tank is empty?
* **The Logic:** You need a variable for `fuel` and a variable for `leak_rate`. Both must be updated every loop iteration.

```python
fuel = 500

usage = 15
leak = 1
leak_growth = 0.10

minutes = 0
current = fuel

while current > 0:
    minutes += 1

    current -= usage
    current -= leak
    leak *= (1 + leak_growth)

print(minutes)
```
