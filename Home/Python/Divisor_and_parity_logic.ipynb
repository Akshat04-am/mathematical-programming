### Challenge 1: The Accumulator
Given $n = 20$, write a script that loops from 1 to $n$. Instead of printing every divisor, create a variable `total` and **sum** all numbers that are divisible by 4. Print only the final sum at the end.

```python
n = 20

lst = []
for test_factor in range(1, n+1):
    if test_factor % 4 == 0:
        lst.append(test_factor)
        print(test_factor)

total = sum(lst)
print(f"The sum is: {total}")
```

### Challenge 2: Prime Detection
A prime number has exactly two factors: 1 and itself. Use a loop to count how many factors a variable $n$ has. After the loop, use an `if` statement to print "Prime" if the count is 2, or "Not Prime" otherwise.

```python
number = 59

lst = []
for test_f in range(1, number+1):
    if number % test_f == 0:
        lst.append(test_f)

if len(lst) == 2:
    print("Prime.")
else:
    print("Not a prime.")
```

### Challenge 3: Shared Divisors
Assign two variables: $a = 24$ and $b = 36$. Write a loop that iterates from 1 up to the smaller of the two numbers. Print every number that is a factor of **both** $a$ and $b$.


```python
a = 24
b = 36

shared_factors = []

for i in range(1, a + 1):
    if a % i == 0 and b % i == 0:
        shared_factors.append(i)

print(shared_factors)
```
