### Euclidian Algorithm and understanding it and code below:

```python
digits = input("Enter a decimal: ")

# Efficiency 1: Accurate exponent
if "." in digits:
    exponent = len(digits.split(".")[1])
else:
    exponent = 0

n = float(digits)
numerator = int(round(n * 10**exponent))
denominator = 10**exponent

# Efficiency 2: Euclidean Algorithm for GCD
a, b = numerator, denominator
while b:
    a, b = b, a % b
gcd = a 

# Simplification
num = numerator // gcd
den = denominator // gcd

print(f"Decimal: {n}")
print(f"Fraction: {num} / {den}")
```
