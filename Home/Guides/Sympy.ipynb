## Loading

```python
import sympy as sp
x = sp.symbols("x")
```

## Basic Functions

```python
# Solve equations
eq = (x-1)*(x+2)*(x-3)
solution = sp.solve(eq,x)

for s in solution:
    print("x = ",s)
```

```python
# Simplify equations
expr = (x + 2)**2 - (x - 2)**2

simplified = sp.simplify(expr)
print(simplified)
```

```python
# Substituting Variables
expr = x**2 - 3*x + 1        
value = int(input("Enter number: "))  

result = expr.subs(x, value)
print("Result:", result)
```

```python
# Substituting Functions
f = x**2 - 3*x + 1
g = x + 1

result = f.subs(x,g)
print(result)
```

## Trig

```python
# simplifying (simplify)
# changing representation (rewrite)
# applying identities (expand_trig)
# simplifying structure (trigsimp)

    # Simplify when want to clean up.
    # Trigsimp when want trig identities applied.
```

```python
# Basic Trignometry. Simplify
expr = sp.cos(x)**2 + sp.sin(x)**2
result = sp.simplify(expr)

print(result)
```

```python
# Expand Trig Identity. Expand_trig
expr = sp.sin(2*x)
expanded = sp.expand_trig(expr)

print(expanded)
```

```python
# Expand trig tdentity and reduce. Expand_trig
expr = sp.tan(x)
rewritten = expr.rewrite(sp.sin)

print(rewritten)

simplified = sp.trigsimp(rewritten)
print(simplified)
```

## LHS = RHS?

```python
# Simple Application
expr1 = (x + 1)**2
expr2 = x**2 + 2*x + 1

check = sp.simplify(expr1 - expr2)
print(check)
```

```python
# Trig Application
expr1 = sp.sin(2*x)
expr2 = 2*sp.sin(x)*sp.cos(x)

check = sp.trigsimp(expr1 - expr2)
print(check)
```
