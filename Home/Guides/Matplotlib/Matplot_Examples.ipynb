### Standard Coordinate Plane Script.

```python
import matplotlib.pyplot as plt
import numpy as np

# Logic Setup
x = np.linspace(-10, 10, 200)
y1, y2 = 3*x, x**3

fig, ax = plt.subplots()

# 1. Automatic reference lines (Zero effort axes)
ax.axhline(0, color='blue', linewidth=1)
ax.axvline(0, color='blue', linewidth=1)

# 2. Data plotting
ax.plot(x, y1)
ax.plot(x, y2)

# 3. Consolidated control (Limits + Grid)
ax.set(xlim=(-10, 10), ylim=(-10, 10))
ax.grid(True, linestyle=':', alpha=0.6)

plt.show()
```

### The Aesthetic "Set and Forget"

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Global style: kills the box, adds grid, sets font
sns.set_theme(style="whitegrid") 

x = np.linspace(-5, 5, 100)
y = np.exp(x)

fig, ax = plt.subplots()
ax.plot(x, y)

# sns.despine() removes the top/right borders in one call
sns.despine(trim=True) 

ax.set(title="Exponential Growth", yscale="log")
plt.show()
```

### The Statistical Fit (Automatic Math)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Synthetic noisy data: y = 2x + noise
x = np.linspace(0, 10, 30)
y = 2 * x + np.random.normal(0, 2, 30)

# One command: scatter plot + regression line + confidence interval
sns.regplot(x=x, y=y, color="teal", marker="+")

plt.show()
```

### The "Grouped Logic" Plot (DataFrame focus)

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Generate data for three different functions
x = np.linspace(0, 10, 50)
df = pd.DataFrame({
    'x': np.tile(x, 3),
    'y': np.concatenate([x, x**1.5, x**2]),
    'type': np.repeat(['linear', 'power_1.5', 'quadratic'], 50)
})

# Automatic legend, colors, and grouping
sns.lineplot(data=df, x='x', y='y', hue='type', style='type')

plt.show()
```
