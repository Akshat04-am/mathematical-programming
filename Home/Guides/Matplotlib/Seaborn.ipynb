### The Only Times You Actually Need Control
Only use these when the **default plot hides the truth**:

* **`ax.set_xlim()` / `ax.set_ylim()`**: Essential because Matplotlib’s auto-scale often fails if you have a single outlier or an asymptote (like $1/x$). It prevents the plot from looking like a flat line.
* **`ax.grid()`**: Essential for visually estimating coordinates without hovering your mouse over every point.
* **`plt.tight_layout()`**: Use it only if your axis labels are getting cut off by the window border.


### 1. The "Global Reset"

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid") # Automatically adds a grid and cleans up borders
```

### 2. The One-Line "Annotation"

```python
import matplotlib.pyplot as plt

# 1. This creates the 'ax' object
fig, ax = plt.subplots() 

# 2. Now you can use it
ax.plot([1, 2, 3], [10, 20, 15], 'ro') 

plt.show()
```

### Examples

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.exp(x)

fig, ax = plt.subplots()
ax.plot(x, y)

# ONLY the essentials:
ax.set_yscale('log') # Logic: Linear scale would hide the curve
ax.grid(True, alpha=0.3)
ax.set_title("Exp Growth") 

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Fix the looks once
sns.set_theme(style="whitegrid")

# 2. Let SNS handle the labels, colors, and legend
sns.lineplot(x=[1, 2, 3], y=[10, 15, 12], label="Trend A")

# 3. Only use PLT for the stuff SNS doesn't touch
plt.title("Fast Plot")
plt.ylim(0, 20)
plt.show()
```
