### All three/four.

```python
import numpy as np
degrees = [0, 15, 30]

# This will fail. Look at the error message (TypeError).
try:
    for i in range(degrees):
        print(i)
except TypeError as e:
    print(f"Error: {e}")
```

```python
degrees = [0, 45, 90]

# A: Direct Grab (Cleanest)
for val in degrees:
    print(f"Value: {val}")
    
# B: Indexed (Needs more typing)
for i in range(len(degrees)):
    print(f"Index: {i} | Value at this index: {degrees[i]}")
```

```python
degrees = [0, 90, 180]
labels = ["Start", "Quarter", "Half"]

# You use range(len()) because you need 'i' to look into TWO different lists
for i in range(len(degrees)):
    deg = degrees[i]
    lab = labels[i]
    print(f"{lab} turn is {deg} degrees.")
```

```python
degrees = [10, 20, 30, 40, 50]

x = len(degrees)
y = range(x)

print(f"The length is a single number: {x}")
print(f"The range is a path object: {y}")
print(f"The path converted to a list: {list(y)}")
```

### Range(len(degree)) and Range(Degree)

###1. The range(len(x)) Approach

```python
# The range(len(x)) Approach
prices = [100, 200, 300]
items = ["Book", "Pen", "Desk"]

# len(prices) becomes 3. range(3) becomes 0, 1, 2.
for i in range(len(prices)):
    print(f"Item {i}: {items[i]} costs {prices[i]}")
```

###2. The range(x) Approach

```python
limit = 5
# This works because limit is a single integer
for i in range(limit):
    print(i)
```

```python
degrees = [0, 90, 180]
# CRASH: TypeError
for i in range(degrees):
    print(i)
```

###3. The for val in x Approach (The Alternative)

```python
degrees = [0, 45, 90]
for deg in degrees:
    print(deg) # Prints 0, 45, 90 directly
```

### Range(len) vs Enumerate

```python
# The "Old" Way (range(len)):
prices1 = [150.2, 151.5, 149.8]

for i in range(len(prices1)):
    print(f"Day {i}: ${prices1[i]}")
```

```python
# The "Modern" Way (enumerate):
for i, val in enumerate(prices1):
    # val is already unpacked and ready to use
    print(f"Day {i}: ${val}")
```
