Nested Logic: Keep your current sin/cos toggle, but add an extra check: if the val is exactly 1.0 or -1.0, print a message next to it saying "PEAK".

```python
# Alternate bw sin and cos.
import numpy as np

degrees = np.arange(0, 361, 15)

for i, deg in enumerate(degrees):
    rad = np.radians(deg)
    
    # 1. First Layer: Toggle the function based on index
    if i % 2 == 0:
        f_type = "sin"
        val = np.sin(rad)
    else:
        f_type = "cos"
        val = np.cos(rad)
    
    # Clean up floating point noise (e.g., 0.999999999 -> 1.0)
    val = np.round(val, 4)
    
    # 2. Nested Layer: Check the VALUE for a "Peak"
    message = ""
    if val == 1.0 or val == -1.0:
        message = " <--- PEAK"
    
    print(f"{deg:<10} | {f_type:<5} | {val:<10.4f}{message}")
```

### Task 1: The Geometry Toggle
You have a list of radii: `radii = [1, 2, 3, 4, 5]`.
***Logic**: Use `enumerate` and `% 2` to alternate between calculating **Area** ($\pi r^2$) and **Circumference** ($2\pi r$).
* **Labeling**: Inside the `if`, set a variable `calc_name = "Area"`. Inside the `else`, set `calc_name = "Circum"`.
* **Goal**: Print a single line at the end of the loop: `Radius: {r} | {calc_name} | {result:.2f}`.

```python
radii = [1,2,3,4,5]

# Used enumerate for index grabbing rather than source value. 
for i, rad in enumerate(radii):

    if i % 2 == 0:
        calc_name = "Area"
        result = np.pi*rad**2
    else:
        calc_name = "Circum"
        result = 2*np.pi*rad
    
    print(f"{rad} | {calc_name} | {result:.2f}")
    
```

### Task 4: The Physics Trend Tracker (State Labeling)
You are tracking altitude over time: `altitudes = [0, 150, 300, 300, 250, 100]`.
* **Setup**: Before the loop, set `prev_alt = 0`.
* **Logic**: Inside the loop, calculate `diff = current_alt - prev_alt`.
* **Labeling**:
    * If `diff > 0`, `status = "CLIMBING"`.
    * Else if `diff < 0`, `status = "DESCENDING"`.
    * Else, `status = "LEVEL"`.
* **State Update**: Use multiple assignment at the end of the loop to update the "previous" value: `prev_alt = current_alt`.
* **Goal**: Print `Alt: {current_alt} | Status: {status}`.

```python
altitudes = [0, 150, 300, 300, 250, 100]

prev_alt = 0
for current_alt in altitudes:
    diff = current_alt - prev_alt

    if diff > 0:
        status = "CLIMBING"
    elif diff < 0:
        status = "DESCENDING"
    else:
        status = "Level"

    print(f"{diff} | {status}")

    # THE CRITICAL MISSING PIECE: Update the memory for the next turn
    prev_alt = current_alt
```

### The "Labeling" Challenge
Since you want to nail this specific centerpiece, here is a task that uses the same logic but adds a **Nested Label**.

**Data**: `points = [[10, 2], [5, 5], [2, 10], [8, 0]]` (These are $x, y$ coordinates).
**Task**:
1.  **Unpack** the coordinates into `x, y` using a loop.
2.  **Structural Label**: Use `enumerate` and `i % 2` to alternate the "Primary Focus":
    * If `i` is even, `focus = "Horizontal"` and `val = x`.
    * If `i` is odd, `focus = "Vertical"` and `val = y`.
3.  **Nested Algebraic Label**: Inside the loop, add an extra check:
    * If the `val` is greater than 7, set a `status` variable to `"HIGH"`.
    * Otherwise, set `status` to `"NORMAL"`.
4.  **Print**: `Row {i} | {focus} | Value: {val} | Status: {status}`.

```python
points = [[10,2], [5,5], [2,10], [8,0]]

for i, (x,y) in enumerate(points):

    if i % 2 == 0:
        Focus = "Horizontal"
        Val = x
    else:
        Focus = "Vertical"
        Val = y

    if Val > 7:
        Status = "HIGH"
    else:
        Status = "Normal"

    print(f"{i} | {Focus} | Value:{Val} | Status: {Status}")
```
