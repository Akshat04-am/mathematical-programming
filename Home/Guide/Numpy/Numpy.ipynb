### Creating Noise

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 50)

# Low noise (Clean data)
y_clean = x + np.random.normal(0, 0.1, len(x))

# High noise (Messy data)
y_messy = x + np.random.normal(0, 2.0, len(x))

plt.scatter(x, y_messy, label="High Noise", alpha=0.5)
plt.scatter(x, y_clean, label="Low Noise")
plt.legend()
plt.show()
```
