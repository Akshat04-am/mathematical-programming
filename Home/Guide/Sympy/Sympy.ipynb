```python
import sympy as sp
x,y = sp.symbols("x y")
```

###  Solve equation.

```python
eq = (x-1)*(x+2)*(x-3)
solution = sp.solve(eq,x)

for s in solution:
    print("x = ",s)
```

### Simplify equations

```python
expr = (x + 2)**2 - (x - 2)**2

simplified = sp.simplify(expr)
print(simplified)
```

### Inputting Variable

```python
expr = x**2 - 3*x + 1        
value = int(input("Enter number: "))  

result = expr.subs(x, value)
print("Result:", result)
```

### Putting one equation into another.

```python
f = x**2 - 3*x + 1
g = x + 1

result = f.subs(x,g)
print(result)
```

###  Checking if Lhs = Rhs

```python
# Simple Application
expr1 = (x + 1)**2
expr2 = x**2 + 2*x + 1

check = sp.simplify(expr1 - expr2)
print(check)
```

### Checking if lhs = rhs trig equations.

```python
expr1 = sp.sin(2*x)
expr2 = 2*sp.sin(x)*sp.cos(x)

check = sp.trigsimp(expr1 - expr2)
print(check)
```

### Solving for 2nd variable

```python
import sympy as sp
x,y = sp.symbols("x y")

eq2 = sp.Eq(2*x + 10,y)

# sol2 = sp.solve(eq2, x)
sol2 = sp.solve(eq2, x)

for s in sol2:
    print("x = ",s)
```

### Two linear Equations

```python
from sympy import *

x,y = symbols('x y')

# Equations set equal to zero
first = 2*x + y - 1 
second = x - 2*y + 7 

# The solution
print(linsolve([first, second], (x, y))) 
```
