### Simplify

```python
import sympy as sp
x,y = sp.symbols("x y")

expr = sp.cos(x)**2 + sp.sin(x)**2
result = sp.simplify(expr)

print(result)
```

### Expand Trig Identity.

```python
expr = sp.sin(2*x)
expanded = sp.expand_trig(expr)

print(expanded)
```

### Expand trig tdentity and reduce. Expand_trig

```python
expr = sp.tan(x)
rewritten = expr.rewrite(sp.sin)

print(rewritten)

simplified = sp.trigsimp(rewritten)
print(simplified)
```
