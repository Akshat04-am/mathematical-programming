#### Confidence Intervals: Seaborn calculates the shadow (error bounds) automatically. In Matplotlib, this requires np.std() and ax.fill_between().

```python
import matplotlib.pyplot as plt
import numpy as np

# Logic Setup
x = np.linspace(-10, 10, 200)
y1, y2 = 3*x, x**3

fig, ax = plt.subplots()

# 1. Automatic reference lines (Zero effort axes)
ax.axhline(0, color='blue', linewidth=1)
ax.axvline(0, color='blue', linewidth=1)

# 2. Data plotting
ax.plot(x, y1)
ax.plot(x, y2)

# 3. Consolidated control (Limits + Grid)
ax.set(xlim=(-10, 10), ylim=(-10, 10))
ax.grid(True, linestyle=':', alpha=0.6)

plt.show()
```
