```python
import numpy as np

# Pick a random number of iterations
iterations = np.random.randint(1, 10) # Randomly chooses 1, 2, 3, 4, or 5

print(f"Loop will run {iterations} times:\n")

for i in range(iterations):
    print(f"Iteration {i}: The total count was {iterations}")
```

```python
import numpy as np

# Create a list (array) of 6 random numbers
random_data = np.random.randint(0, 101, 6)

counter = 0  # Initialize manual counter
for val in random_data:

    if counter % 2 == 0:
        print(f"Index {counter}: {val}")
    
    counter += 1  # Increment manual counter
```
