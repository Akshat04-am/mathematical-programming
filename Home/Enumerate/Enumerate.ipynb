```python
# Manual counter (Inefficient)
i = 0
for deg in [0, 15, 30]:
    print(i, deg)
    i += 1

# Using enumerate (Standard)
for i, deg in enumerate([0, 15, 30]):
    print(i, deg)
```

```python
degrees = [0, 15, 30, 45]

for i, deg in enumerate(degrees):
    # i is the "count" (0, 1, 2, 3)
    # deg is the "value" (0, 15, 30, 45)
    
    if i % 2 == 0:
        print(f"Index {i} is even: calculating sin({deg})")
    else:
        print(f"Index {i} is odd: calculating cos({deg})")
```

```python
data = [10, 20, 30, 40]

i = 0  # Initialize the index manually
for x in data:
    print(f"Step {i}: Value is {x}")
    
    # Manual increment is required
    i += 1
```

Index Multiplier: Create a list [2, 4, 6, 8]. Use a manual counter i to print the result of the current number multiplied by its index (e.g., $2 \times 0$, $4 \times 1$, etc.).

```python
import numpy as np

x = np.arange(0,101,17)

counter = 0
for val in x:

    result = val * counter
    print(f"Index {counter} | Value {val} | Result: {result}")
    counter += 1

for i, deg in enumerate(x):
    print(f"Index {i} | Value {deg}")
```
